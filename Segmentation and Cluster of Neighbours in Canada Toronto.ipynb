Fetching the Data from Wiki page using BeautifulSoup.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
resp = requests.get(url).text
soup = BeautifulSoup(resp,'lxml')
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
Converting it into a dataframe using pandas library
```

```python
data_from_table = pd.read_html(tab)
df = data_from_table[0]
print("Before removing the Not Assigned Boroughs the shape of df is {}".format(df.shape))
df = df[df['Borough'] != 'Not assigned']
print("after removing the Not Assigned Boroughs the shape of df is {}".format(df.shape))
df.head()
```

Grouping Boroughs, cleaning by removing not assigned in Neighborhood column  and displaying Shape of Data Frame.

```python
df1 = df.groupby(['Postal code','Borough'], sort=False ).agg(" , ".join)
df1.reset_index(inplace=True)
# Replacing / of the neighbourhoods with ,
df1['Neighborhood'] = df1['Neighborhood'].str.replace('/',',')
# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df1['Neighborhood'] = np.where(df1['Neighborhood'] == 'Not assigned',df1['Borough'],df1['Neighborhood'])
print("Dimensions of Data Frame after cleaning and grouping : {}".format(df1.shape))
print("-----------------------------Sneak Peak Of Data ------------------------------------")
print(df1.head())
```

Fetching Latitude and Longitude details for All 

```python
location_df = pd.read_csv('http://cocl.us/Geospatial_data')
location_df.head()
```

changing name of Postal Code in location_df to "Postal code"

```python
#renaming column
location_df.rename(columns={"Postal Code":"Postal code"}, inplace=True)
#joining df1 and location_df
actual_df = pd.merge(df1,location_df,on='Postal code')
actual_df.head()
```

Now we need only Toronto neighborhood details for visualization

```python
toronto_df = actual_df[actual_df['Borough'].str.contains('Toronto')]
toronto_df
```

Display on Map - Toronto

```python
toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
for lat,lng,borough,neighbourhood in zip(toronto_df['Latitude'],toronto_df['Longitude'],toronto_df['Borough'],toronto_df['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='yellow',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(toronto_map)
toronto_map
```

Clustering the Data frame using K means.

```python
toronto_clustering = toronto_df.drop(['Postal code','Borough','Neighborhood'],1)
km_clustering=KMeans(n_clusters=4,random_state=0).fit(toronto_clustering)
km_clustering.labels_
toronto_df.insert(0,'Cluster',km_clustering.labels_)
```
